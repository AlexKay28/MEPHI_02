# Task 1

При уровне значимости **alpha = 0.1** проверить гипотезу о равенстве дисперсий двух нормально распределенных случайных величин X и Y на основе выборочных данных (табл.1) при альтернативной гипотезе .

- H0: X.std == Y.std
- H1: X.std != Y.std

```python
df1 = pd.DataFrame()
df1['X_vals']   = [6.1, 6.5, 6.6, 7.0, 7.4]
df1['X_counts'] = [2,3,1,4,2]
df1['Y_vals']   = [5.8, 6.0, 6.2, 6.3, 6.8]
df1['Y_counts'] = [6,4,5,2,3]

df1['X'] = np.array([df1.iloc[i]['X_vals']*df1.iloc[i]['X_counts'] for i in range(df1.shape[0])])
df1['Y'] = np.array([df1.iloc[i]['Y_vals']*df1.iloc[i]['Y_counts'] for i in range(df1.shape[0])])

X_std = df1['X'].std()
Y_std = df1['Y'].std()
F_experimental = X_std/Y_std

fx = 12 - 1
fy = 20 - 1
F_theoretical = 2 # (df2 : df1) = (19 : 11) F_theor -> ~2

print(f'std X = {X_std:.3f}') # unbiased
print(f'std Y = {Y_std:.3f}')
print(f'F_experimental = {F_experimental:.3f}')
print(f'F_theoretical  = {F_theoretical:.3f}')

if F_experimental < F_theoretical:
    print('Нулевая гипотеза принимается, альтернативная гипотеза отклоняется')
else:
    print('Нулевая гипотеза отклоняется, принимается альтернативная гипотеза')
    
df1
```

# Task 2

При уровне значимости **alpha = 0.05** методом дисперсионного анализа проверить нулевую гипотезу о влиянии фактора на качество объекта на основании пяти измерений для трех уровней фактора (табл. 2).

- H0: F_exp <= F_theor

- H1: F_exp  > F_theor

```python
df2 = pd.DataFrame()

df2['F1'] = [9, 11, 10, 12, 9]
df2['F2'] = [4, 6, 5, 6, 5]
df2['F3'] = [12, 18, 24, 20, 23]

# экземпляр фрейма с данными из показательной задачи лекции 12-13
# df2['F1'] = [600, 420, 510, 435, 495]
# df2['F2'] = [570, 450, 630, 450, 450]
# df2['F3'] = [690, 570, 600, 570, 600]
# df2['F4'] = [450, 510, 450, 510, 540]

df2
```

```python
# Cреднее по столбцам
columns = [column for column in df2.columns]
for column in columns:
    print(f'Mean {column} = ', df2[column].mean(), end = '\t\t ')
print()

# Oпределяем оценки дисперсий для каждого столбца: 
columns_std = []
for column in columns:
    columns_std.append(df2[column].std())
    print(f'Std  {column} = ', round(df2[column].std(), 3), end = '\t ')
print()
print()

# Oпределяем расчетное значение критерия Кохрена: 
G_Kohren = max(columns_std)/sum(columns_std)
print(f'G_Kohren = {G_Kohren:.3f}', end=' | ')


# Oпределяем табличное значение G-критерия по таблице приложения
G_theoretical = 0.7457 # G(alpha = 0.5, n=3, f=5-1)\
print(f'G_theoretical = {G_theoretical:.3f}')

if G_Kohren < G_theoretical:
    print('G_Kohren < G_theoretical')
    print('! оценки дисперсий однородны и можно приступить к проведению ДА')
else:
    raise NameError('Дисперсионный анализ невозможен!\n\t   G_Kohren > G_theoretical')
print()

# Подсчитываем итоги по столбцам: 
column_sums = []
for column in columns:
    column_sums.append(df2[column].sum())
    print(f'Column {column} =', df2[column].sum(), end = '\t\t')

print()

n = df2.shape[0]
m = df2.shape[1]

# Вычисляем сумму квадратов всех наблюдений: 
df2_2 = df2**2
Q1 = df2_2[df2_2.columns].sum(axis=1).sum()
print(f'Q1 = {Q1:.2f}')

# Вычисляем сумму квадратов итогов по столбцам, деленную на число наблюдений в столбце: 
Q2 = sum([column_sum**2 for column_sum in column_sums])/n
print(f'Q2 = {Q2:.2f}')

# Рассчитаем квадрат общего итога, деленный на число всех наблюдений: 
Q3 = sum([column_sum for column_sum in column_sums])**2/(m*n)
print(f'Q3 = {Q3:.2f}')
print()

# Определяем сумму квадратов для столбца: 
SSx = Q2 - Q3
print(f'SSx = {SSx:.3f}')

# Определяем общую сумму квадратов: 
SSm = Q1 - Q3
print(f'SSm = {SSm:.3f}')

# Определяем остаточную сумму квадратов для оценки ошибки эксперимента:
SSe = Q1 - Q2
print(f'SSe = {SSe:.3f}')
print()

# Рассчитываем оценку диспресии Sx^2
Sx2 = SSx/(m-1)
print(f'Sx2 = {Sx2:.3f}')

# Рассчитываем оценку дисперсии Se^2
Se2 = SSe/(m*(n-1))
print(f'Se2 = {Se2:.3f}')
print()

# Вычисляем расчетное значение F-критерия 
F_exp = Sx2/Se2
print(f'F расчетное \t= {F_exp:.5f}')

# Определяем теоретическое значение F-критерия (alpha =0.5, f1=2, f2=12)
F_theor = 3.88530
print(f'F теоретическое = {F_theor:.5f}')

if F_exp < F_theor:
    print('Нулевая гипотеза принимается, альтернативная гипотеза отклоняется')
else:
    print('Нулевая гипотеза отклоняется, принимается альтернативная гипотеза')

# Таким образом, влияние факторов на зависимую величину 
# следует признать значимым. 

# Отклоняем нулевую гипотезу!
```

# Answers

1. Нулевая гипотеза принимается, альтернативная гипотеза отклоняется
2. Нулевая гипотеза отклоняется, принимается альтернативная гипотеза
