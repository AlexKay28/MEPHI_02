```python
data_higgs = pd.read_csv('higgs_reduced.csv')
```

```python
data_higgs.head()
```

```python
data_higgs.rename({'0':'labels'}, axis=1, inplace=True)
data_higgs.head()
```

```python
feat_name = 'lepton pT, lepton eta, lepton phi, missing energy magnitude, missing energy phi, jet 1 pt, jet 1 eta, jet 1 phi, jet 1 b-tag, jet 2 pt, jet 2 eta, jet 2 phi, jet 2 b-tag, jet 3 pt, jet 3 eta, jet 3 phi, jet 3 b-tag, jet 4 pt, jet 4 eta, jet 4 phi, jet 4 b-tag, m_jj, m_jjj, m_lv, m_jlv, m_bb, m_wbb, m_wwbb'
feat_names_list = feat_name.split(', ')
feat_names_list
```

```python
def create_dict_names(nms):
  num_feat = len(nms)
  dict_keys = []
  for i in range(1, num_feat+1):
    dict_keys.append(str(i))
  dict_names = dict(zip(dict_keys, nms))
  return dict_names
```

```python
dict_col = create_dict_names(feat_names_list)
dict_col
```

```python
data_higgs.rename(dict_col, axis=1, inplace=True)
data_higgs.head()
```
